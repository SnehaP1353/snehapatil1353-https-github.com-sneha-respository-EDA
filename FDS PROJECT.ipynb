## 1.Introduction of DataSet:

```python
The The Wine Quality (Red) dataset is a popular and widely used dataset in the field
of data analysis and machine learning. It is specifically curated for the analysis of red wine quality based on various 
physicochemical properties. This dataset is often employed by researchers, data scientists, and enthusiasts to explore 
relationships between different attributes of red wine and its overall quality.
```

## 2.Description of DataSet:

```python
DESCRIPTION OF DATASET : Here is the detailed description of the Wine-Quality dataset is.
     is a detailed description of the dataset:
*Attributes:
1)Fixed Acidity: The amount of non-volatile acids in the wine.
2)Volatile Acidity: The amount of acetic acid in the wine, which can contribute to the taste of vinegar.
3)Citric Acid: The amount of citric acid in the wine, which can add freshness and flavor.
4)Residual Sugar: The amount of sugar remaining after fermentation stops. Wines with higher residual sugar are sweeter.
5)Chlorides: The amount of salt in the wine.
6)Free Sulfur Dioxide: The free form of SO2 exists in equilibrium between molecular SO2 (dissolved gas) and bisulfite ion.
7)Total Sulfur Dioxide: The total amount of sulfur dioxide, including both free and bound forms. It is used as an antimicrobial
    and antioxidant.
8)Density: The density of the wine, which is generally related to the alcohol content.
9)pH: The pH level measures the acidity or basicity of the wine on a scale of 0 to 14.
10)Sulphates: The amount of sulfates in the wine. Sulfates are additives that can contribute to preserving the wine.
11)Alcohol: The alcohol content of the wine.

*Target Variable:
1)Quality: The quality of the wine, typically rated on a scale from 3 to 8, with higher values indicating better quality.
2)Data Format:
The dataset is usually provided in a tabular format, with each row representing a specific red wine sample and each column 
representing a different attribute.
The data may be in CSV (Comma-Separated Values) format or another standard tabular format.
3)Size:
The dataset typically contains several hundred instances, providing enough data for meaningful analysis.
4)Origin:
The dataset is often used as a benchmark dataset in machine learning and data analysis courses. Its origin can be traced to
various machine learning repositories and research studies.
5)Usage:
Researchers and data scientists use this dataset for tasks such as predictive modeling, feature selection, and understanding the
relationships between different attributes and wine quality.
6)Missing Values:
It's common to find datasets with complete information, but analysts should always check for and handle any missing values if 
present.
```

```python
#histogram for a specific feature
plt.figure(figsize=(10,7))
plt.hist(df["chlorides"])
plt.xlabel("chlorides")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Comparison between quality based on alcohol and sulphates')
sns.scatterplot(x=df['alcohol'],y=df['sulphates'],hue=df['quality'],s=50)
```

```python
plt.title("residual sugar")
sns.boxplot(df['residual sugar'],color='orange')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.scatter3D(df['sulphates'],df['citric acid'],df['quality'],edgecolor='black');
ax.set_xlabel("sulphates")
ax.set_ylabel("citric acid")
ax.set_zlabel("quality")
```
